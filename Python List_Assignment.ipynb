### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
# Please write The solution Here for above question 7:
n=[134,121,155]
temp=n
rev=0

while(n>0):
   dig=n%10
   rev=rev*10+dig
   n=n//10

if(temp==rev):
   print("Yes, the number is a palindrome!")
else:
   print("No, the number isn't a palindrome!")
```

### Question 8: Write a Python program to find the common elements between two lists.

- `sample input : `
    - `var1 = [1,2,3,4,10]`
    - `var2 = [10,11,12,14]`
- `sample output :`
    - `result : 10`

```python
# Please write The solution Here for above question 8:
list1=[1,2,3,4,10]
list2=[10,11,12,14]
a=set(list1)
print(a)
b=set(list2)
print(b)

c=a.intersection(b)
print(list(c))





```

```python
var1=[1,2,3,4,10]
var2=[10,11,12,14]
def cmnset(var1,var2):
  var1s=set(var1)
  var2s=set(var2)

  if len(var1s.intersection(var2s))>0:
    print(var1s.intersection(var2s))
  else:
    print('no common element in list')
cmnset(var1,var2)

```
